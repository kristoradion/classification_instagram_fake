```python
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

def evaluate_classification_algorithms(df, target_variable, algorithms, num_folds=4):
    # Extract features (X) and the target variable (y)
    X = df.drop(columns=[target_variable])
    y = df[target_variable]

    results = {}  # Dictionary to store results for each algorithm

    for algorithm in algorithms:
        if algorithm == 'RandomForest':
            clf = RandomForestClassifier()
        elif algorithm == 'LogisticRegression':
            clf = LogisticRegression()
        elif algorithm == 'KNeighbors':
            clf = KNeighborsClassifier()
        elif algorithm == 'SVM':
            clf = SVC()
        else:
            raise ValueError(f"Unsupported algorithm: {algorithm}")

        # Perform cross-validation and calculate accuracy
        accuracy_scores = cross_val_score(clf, X, y, cv=num_folds, scoring='accuracy')

        # Calculate the mean accuracy score
        mean_accuracy = accuracy_scores.mean()

        results[algorithm] = mean_accuracy

    return results

# Example usage:
# Load your DataFrame and specify the target variable
# Replace 'your_dataframe.csv' and 'target_variable' with your actual data
algorithms_to_try = ['RandomForest', 'LogisticRegression', 'KNeighbors', 'SVM']
results = evaluate_classification_algorithms(df, "class", algorithms_to_try, num_folds=4)
print(results)
```
